## Averaging 平均平滑
计算卷积框覆盖区域所有的相符的平均值得到的卷积结果

## Gassian 高斯模糊
把卷积核换成高斯核（原来的每个方框的志是相等的，现在里面的值符合高斯分布，方框中心最大，其余根据距离中心的元素递减，构成一个高斯小山包，原来的求平均数变成加权平均，权就是方框里的值）

## Median 中值模糊
给偶偶名思议就是与卷积框对应像素的中值来替代中心像素的值

## Bilateral 双边滤波
能在保持边界清晰的情况下有效的去除噪音。
双边滤波在同时使用空间高斯权重和灰度值相似性高斯权重，空间高斯函数确保只有紧邻区域的像素对中心有影响，灰度值相似性高斯函数确保只有与中心像素灰度值相似的才会被用来做模糊计算，这种方法确保边界不会被模糊掉。因为边界处灰度值变化比较大。

```python
import cv2 as cv
import matplotlib.pyplot as plt
from utils.im import *

kernelsize = [(3,3),(9,9),(15,15)]
plt.figure(figsize=(15,10))


image = cv.imread("./images/tree.jpeg")
plt.subplot(2, 2, 1);
plt.axis("off")
plt.title("original")
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img)

for i, kernel in enumerate(kernelsize):
    plt.subplot(2, 2, i + 2)
    # 平均平滑
    blur = cv.blur(image, kernel)
    img = cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)

    # 不显示坐标
    plt.axis("off")
    # 设置字体
    plt.title("Blurred" + str(kernel))
    plt.imshow(img)
plt.show()
```

```python
# img = cv.imread("./images/imagenet/elephant.jpg")
params = [(11,21,7),(11,41,21),(15,75,75)]

plt.figure(figsize=(15,10))

image = cv.imread("./images/tree.jpeg")
plt.subplot(2, 2, 1);
plt.axis("off")
plt.title("original")
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img)

# 邻域直径，灰度值相似性高斯函数标准差，空间高斯函数标准差
for i, (diameter, sigmaColor, sigmaSpace) in enumerate(params):
    plt.subplot(2, 2, i + 2)
    blur = cv.bilateralFilter(image, diameter, sigmaColor, sigmaSpace)
    img = cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)
    plt.axis("off")
    plt.title("Blurred" + str((diameter, sigmaColor, sigmaSpace)))
    plt.imshow(img)
plt.show()
```
