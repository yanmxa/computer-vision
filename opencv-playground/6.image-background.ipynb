```python
# coding=utf-8
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def showBGR(img):
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(imgRGB)
    plt.show()
```

```python
img = cv2.imread('./images/yan.jpg')
```

```python
# 
rows,cols,channels = img.shape
# img=cv2.resize(img,None,fx=0.5,fy=0.5)
rows,cols,channels = img.shape
# cv2.imshow('img',img)
showBGR(img)
```

```python
# hsv
hsv=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
lower_blue=np.array([90,70,70])
upper_blue=np.array([110,255,255])
mask = cv2.inRange(hsv, lower_blue, upper_blue)
```

```python
showBGR(mask)
```

```python
erode=cv2.erode(mask,None,iterations=1)
showBGR(erode)
```

```python
dilate=cv2.dilate(erode,None,iterations=1)
# cv2.imshow('dilate',dilate)
showBGR(dilate)
```

```python
#
for i in range(rows):
    for j in range(cols):
        if dilate[i,j]==255:
            img[i,j]=(0,0,255)#BGR
showBGR(img)
```

```python

blur = cv2.blur(img, (5, 5))
showBGR(blur)
```

```python
blur = cv2.GaussianBlur(img, (3, 3), 0)
showBGR(blur)
```

```python
#
for i in range(rows):
    for j in range(cols):
        if dilate[i,j]==255:
            img[i,j]=(0,0,255)#BGR
            
# 
# img1 = cv2.resize(img, (358,441), interpolation=cv2.INTER_CUBIC)
# img2 = cv2.resize(blur, (358,441), interpolation=cv2.INTER_CUBIC)
img1 = img
blur = cv2.GaussianBlur(img, (3, 3), 0)

img2 = blur

# 
cv2.imwrite("./images/yan_red_1cun.jpeg", img1, [int( cv2.IMWRITE_JPEG_QUALITY), 35])
cv2.imwrite("./images/yan_red_1cun_blur.jpeg", img2, [int( cv2.IMWRITE_JPEG_QUALITY), 35])
```
