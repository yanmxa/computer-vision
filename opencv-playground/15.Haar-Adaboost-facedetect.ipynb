```python
from utils.im import *
import cv2 as cv
import numpy as np
```

```python
kun = cv.imread("./images/kun.jpeg")
meng = cv.imread("./images/yan.jpg")
show(kun)
show(meng)
```

```python
# 级联分类器
detector = cv.CascadeClassifier('haarcascade_frontalface_alt2.xml') # 特征分类器
rects = detector.detectMultiScale(meng,                  # 输入图片
                                  scaleFactor=1.1,        # 每次图像缩小的比例
                                  minNeighbors=3,        # 匹配成功所需要周围的矩形数目
                                  minSize=(10,10),        # 匹配人脸的最小范围10*10 px
                                  flags=cv.CASCADE_SCALE_IMAGE)
# rects = detector.detectMultiScale(kun,scaleFactor=1.1,minNeighbors=2,minSize=(10,10),flags=cv.CASCADE_SCALE_IMAGE)
for (x, y, w, h) in rects:
    cv.rectangle(meng, (x, y), (x + w, y + h), (0,255,0), 2)
    
show(meng)
```

```python
def facedetect(image):
    image = cv.imread(image)
    detector = cv.CascadeClassifier('haarcascade_frontalface_alt2.xml')
    rects = detector.detectMultiScale(image,
                                      scaleFactor=1.1,
                                      minNeighbors=2,
                                      minSize=(10,10),
                                      flags=cv.CASCADE_SCALE_IMAGE)
    for (x, y, w, h) in rects:
        cv.rectangle(image, (x, y), (x + w, y + h), (0,255,0), 2)
    show(image)
```

```python
facedetect("./images/people.jpg")
```
