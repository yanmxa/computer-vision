# OpenCV 图像读取显示与存储

1. 实验目标
    - OpenCV读入图像、显示图片、保存图像
    - 主要函数：cv.imread(), cv.imshow(), cv.imwrite()
    - 使用matplotlib显示图片
2. 前置条件
    - python
    - Numpy
    - Matplotlib
    - OpenCV

```python
import numpy as np
import cv2 as cv
```

## 1. 图片读取 cv.imread()
- 参数1：图片路径
- 参数2：
    - 1 cv.IMREAD_COLOR, 读入彩色图片，透明度会被忽略
    - 0 cv.IMREAD_GRAYSALE, 以灰度模式读入图片
    - -1 cv.IMREAD_UNCHNAGED, 读入一幅图片，包好图像的alpha通道

```python
img = "/Users/meng.yan/workspace/opencv-playground/images/kun.jpeg"
images_dir = "/Users/meng.yan/workspace/opencv-playground/images/"
img1 = cv.imread(img, 0)
img2 = cv.imread(img, 1)
img3 = cv.imread(img, -1)
```

```python
print(img1.shape)
print(img2.shape)
print(img3.shape)
type(img1)
```

## 2.显示图片
- 方法一：cv.imshow(), 使用窗口显示图像，窗口大小自动适应图片大小
- 方法二：cv.namedWindow先创建窗口
    - 参数： cv.WINDOW_AUTOSIZE 自动生成窗口大小，cv.WINDOW_NORMAL标签允许调整窗口大小
- 方法三： matplotlib显示

```python
cv.imshow("img1", img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
cv.namedWindow("image", cv.WINDOW_NORMAL)
cv.imshow("image2", img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(img2)
```

### 这里颜色出现了问题，是因为cv.imread()读出爱的图像是GBR顺序存储,而matplotlib是RGB的顺序，所以需要cv.cvtColor函数转换一下。

```python
imageRGB = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
plt.imshow(imageRGB)
```

## 3. 保存图像
- cv.imwrite()来保存一个图像
- 确定文件名字

```python
cv.imwrite(images_dir + "img1.png", img1)
cv.imwrite(images_dir + "img2.png", img2)
```
