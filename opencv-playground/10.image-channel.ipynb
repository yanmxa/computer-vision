```python
image = cv.imread('./images/tree.jpeg')
M = np.ones(image.shape, dtype='uint8')*100
image = cv.add(image, M)
show(image)
```

```python
image = cv.imread('./images/tree.jpeg')
M = np.ones(image.shape, dtype='uint8')*100
image = cv.subtract(image, M)
show(image)
```

```python
image = cv.imread('./images/tree.jpeg')
image.shape
```

```python
(B, G, R) = cv.split(image)
```

```python
b = cv.merge([B, 
              np.zeros(image.shape[:2], dtype='uint8'), 
              np.zeros(image.shape[:2], dtype='uint8')])
show(b)
```

```python
g = cv.merge([np.zeros(image.shape[:2], dtype='uint8'), 
              G, 
              np.zeros(image.shape[:2], dtype='uint8')])
show(g)
```

```python
r = cv.merge([np.zeros(image.shape[:2], dtype='uint8'), 
              np.zeros(image.shape[:2], dtype='uint8'), 
              R])
show(r)
```

```python
img = cv.merge([B, G, R])
show(img)
```

```python
cv.imshow("R", R)
cv.imshow("B", B)
cv.imshow("G", G)
cv.waitKey(0)
cv.destroyAllWindows()
```
